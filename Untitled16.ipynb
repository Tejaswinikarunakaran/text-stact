```python
import matplotlib.pyplot as plt

# x-coordinates of left sides of bars
left = [1, 2, 3, 4, 5]

# heights of bars
height = [10, 24, 36, 40, 5]

# labels for bars
tick_label = ['2016', '2017', '2018', '2019', '2020']

# plotting a bar chart
plt.bar(left, height, tick_label = tick_label,
        width = 0.8, color = ['blue', 'green'])

# naming the x-axis
plt.xlabel('price')
# naming the y-axis
plt.ylabel('sales')
# plot title
plt.title('influnce of sales')

# function to show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# x axis values
x = [10,20,30,40,50,60]
# corresponding y axis values
y = [3,5,7,9,11,13]

# plotting the points
plt.plot(x, y, color='pink', linestyle='dashdot', linewidth = 3,
         marker='o', markerfacecolor='blue', markersize=12)

# setting x and y axis range
plt.ylim(3,15)
plt.xlim(10,60)

# naming the x axis
plt.xlabel('x - axis')
# naming the y axis
plt.ylabel('y - axis')

# giving a title to my graph
plt.title('Some cool customizations!')

# function to show the plot
plt.show()
```
