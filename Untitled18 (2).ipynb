```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 6, 10)
y = x**2
```

```python
x
```

```python
y
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.xlabel('X values')
plt.ylabel('Y values')
plt.title('Graph of $y = x^2$')

plt.plot(x, y)
plt.show()
```

```python
x1 = np.linspace(-np.pi, np.pi, 256)
x2 = np.linspace(-np.pi, np.pi, 256)

y1 = np.cos(x1)
y2 = np.sin(x2)
```

```python
plt.plot(x1, y1, x2, y2)
```

```python
plt.figure(figsize = (10, 6), dpi = 70)

plt.plot(x1, y1)
plt.plot(x2, y2)

plt.show()
```

```python
plt.figure(figsize = (10, 6), dpi = 80)

plt.plot(x1, y1, 'r-.')
plt.plot(x2, y2, 'b--')

plt.show()
```

```python
plt.figure(figsize = (10, 6), dpi = 80)

plt.plot(x1, y1, color = 'purple')
plt.plot(x2, y2, color = 'green')

plt.show()
```

```python
plt.figure(figsize = (10, 6), dpi = 80)

plt.plot(x1, y1, color = 'purple', lw = 5.5)
plt.plot(x2, y2, color = 'green', lw = 2.5)

plt.show()
```

```python
plt.figure(figsize = (10, 6), dpi = 80)

plt.plot(x1, y1, color = 'purple', lw = 2.5, ls = '--')
plt.plot(x2, y2, color = 'green', lw = 2.5, ls = '-.')

plt.show()
```

```python
plt.figure(figsize = (10, 6), dpi = 80)

plt.plot(x1, y1, color = 'purple', lw = 2.5, ls = '--')
plt.plot(x2, y2, color = 'green', lw = 2.5, ls = '-.')

plt.show()
```

```python
plt.figure(figsize = (10, 6), dpi = 80)

plt.plot(x1, y1, color = 'purple', lw = 2.5, ls = '--')
plt.plot(x2, y2, color = 'green', lw = 2.5, ls = '-.')

plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi])
plt.yticks([-1, 0, 1])

plt.show()
```

```python
plt.figure(figsize = (10, 6), dpi = 80)

plt.plot(x1, y1, color = 'purple', lw = 2.5, ls = '--')
plt.plot(x2, y2, color = 'green', lw = 2.5, ls = '-.')

plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi])
plt.yticks([-1, 0, 1])

plt.show()
```
